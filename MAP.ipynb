```python
ladakh_row = {'ST_NM': 'Ladakh', 'INCIDENT_EVENT_ID': None, }
data = data.append(ladakh_row, ignore_index=True)
```

```python
# Loading geo json to plot maps
url = "https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson"
with request.urlopen(url) as f:
    geo_india = json.load(f)

geo_df = gpd.read_file(url)
geo_df["lon"] = geo_df["geometry"].centroid.x
geo_df["lat"] = geo_df["geometry"].centroid.y
geo_df['ST_NM'] = geo_df['ST_NM'].str.title()
print(geo_df['ST_NM'])
```

```python
# Calculate state-wise event counts
statewise_event_counts = data['ST_NM'].value_counts().reset_index()
print(statewise_event_counts, statewise_event_counts.values)
statewise_event_counts.columns = ['admin1', 'count']

# Conditionally assign the count for Ladakh as 0
for index, row in statewise_event_counts.iterrows():
    if row['admin1'] == 'Ladakh':
        statewise_event_counts.loc[index, 'count'] = 0

print(statewise_event_counts)


for i in statewise_event_counts['admin1']:
    if i == 'JAMMU AND KASHMIR':
        print("hey")
        statewise_event_counts['admin1'].replace(i, "Jammu & Kashmir", inplace=True)

statewise_event_counts['admin1'] = statewise_event_counts['admin1'].str.title()
# print(statewise_event_counts['admin1'])
```

```python
# merging
merged_data = geo_df.merge(statewise_event_counts, left_on='ST_NM', right_on='admin1')
print(merged_data)
```

```python
import plotly.graph_objects as go
import json
import zipfile
import os

# Your existing code goes here

# Create a directory to store HTML files
output_directory = '/content/html_files/'
os.makedirs(output_directory, exist_ok=True)

# Save each state as an HTML file
for state in merged_data['admin1'].unique():
    state_data = merged_data[merged_data['admin1'] == state]

    # Identify the state's latitude and longitude boundaries
    state_lon_min, state_lon_max = state_data['lon'].min(), state_data['lon'].max()
    state_lat_min, state_lat_max = state_data['lat'].min(), state_data['lat'].max()

    state_html_path = os.path.join(output_directory, f"{state}_map.html")  # Set HTML file path for each state

    state_fig = go.Figure(data=go.Choropleth(
        geojson=geo_india,
        featureidkey='properties.ST_NM',
        locationmode='geojson-id',
        locations=state_data['admin1'],
        z=state_data['count'],
        colorscale=[[0, 'white'], [1, 'red']],
        autocolorscale=False,
        marker_line_color='grey',
        marker_opacity=0.7,
        colorbar=dict(
            title={'text': "Number of Events"},
            thickness=15,
            len=0.5
        )
    ))

    state_fig.update_traces(showscale=False)

    state_fig.update_geos(
        lonaxis_range=[state_lon_min - 1, state_lon_max + 1],  # Adjust the longitude range
        lataxis_range=[state_lat_min - 1, state_lat_max + 1]   # Adjust the latitude range
    )

    state_fig.update_layout(
        title_text=f"{state} - Riot Occurrences",
        geo=dict(
            scope='asia',
            projection_scale=0.2,  # Adjust this value for less zoom
            bgcolor='white',
            showcountries=True  # Show border lines for states
        )
    )

    state_fig.write_html(state_html_path)  # Save the HTML file with the correct path

# Create a zip file containing all HTML files
zip_file_path = '/content/html_files.zip'
with zipfile.ZipFile(zip_file_path, 'w') as zipf:
    for root, dirs, files in os.walk(output_directory):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, os.path.relpath(file_path, output_directory))

# Display the zip file path (optional)
print("HTML files saved in zip:", zip_file_path)


```

```python
import plotly.graph_objects as go
import plotly.io as pio

# Your existing code goes here

# Create the choropleth map using plotly
fig = go.Figure(data=go.Choropleth(
    geojson=geo_india,
    featureidkey='properties.ST_NM',
    locationmode='geojson-id',
    locations=merged_data['ST_NM'],
    z=merged_data['count'],
    text=merged_data['admin1'],
    autocolorscale=False,
    colorscale='oranges',
    marker_line_color='grey',
    colorbar=dict(
        title={'text': "Number of Events"},
        thickness=15,
        len=0.5
    )
))

fig.update_geos(
    visible=False,
    projection=dict(
        type='conic conformal',
        parallels=[12.472944444, 35.172805555556],
        rotation={'lat': 24, 'lon': 80}
    ),
    lonaxis={'range': [68, 98]},
    lataxis={'range': [6, 38]}
)

fig.add_trace(go.Scattergeo(
    lon=merged_data['lon'],
    lat=merged_data['lat'],
    mode='text',
    text=['{}<br>{}'.format(k, v) for k, v in zip(merged_data['admin1'].str.title(), merged_data['count'])],
    textfont={
        'color': 'black',
        'size': 9
    },
    name='State Wise Event Count',
))

fig.update_layout(
    title=dict(
        text="India - Region wise Riot Occurrences",
        xanchor='center',
        x=0.5,
        yref='paper',
        yanchor='bottom',
        y=1,
        pad={'b': 10},
    ),
    dragmode=False,
    margin={'r': 0, 't': 30, 'l': 0, 'b': 0},
    height=700
)

# Save the map as an HTML file
html_file_path = "choropleth_map.html"
pio.write_html(fig, file=html_file_path)

# Open the saved HTML file in a web browser
import webbrowser
webbrowser.open(html_file_path)
```
